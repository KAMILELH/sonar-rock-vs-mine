importing the dependecies

```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Loading the DataSet to panda dataframe
sonar_dataset = pd.read_csv("sonar dataset.csv", header=None)
```

```python
sonar_dataset.head();
```

```python
sonar_dataset.shape
```

```python
sonar_dataset.describe();   #donne les statistiques de donnees 
```

```python
sonar_dataset[60].value_counts()
```

```python
# separer target(R et M) et les donnees
X = sonar_dataset.drop(columns = 60, axis = 1)
Y = sonar_dataset[60]
```

classer les donnees en Test et Train 

```python
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size=0.1, stratify=Y, random_state=1)

#stratify: les targets sont divise egalement entre testData et trainData
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
# entrainer le model de logistic regression avec Trainning Data
model.fit(X_train, Y_train)
```

Evaluation de model

```python
#accuracy on trainning Data
X_train_prediction = model.predict(X_train)
train_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print(train_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print(test_data_accuracy)  #faut augmenter dataset   
```

Creer le Systeme de Prediction

```python
input_data = ()

#changing input to np array
input_data_as_array = np.asarray(input_data)

#reshape np array as we're predicting for one instance
input_data.reshape = input_data_as_array.reshape(1,-1)

prediction = model.predict(input_data.reshape)
print(prediction)



```
